```python
from pydantic import BaseModel,StrictStr,StrictInt
```

```python
from dataclasses import dataclass

@dataclass
class Person():
    name:str
    age:int
    city:str

person=Person(name=11,age=16,city="Srinivas_nagar")
print(person)
```

```python
person=Person(name="Soni",age=21,city="Srinivas_nagar")
print(person)
```

```python
class Person1(BaseModel):
    name:StrictStr
    age:StrictInt
    city:StrictStr

person=Person1(name=11,age=15,city="Srinivas nagar")
print(person)
```

```python
from typing import Optional
class Employee(BaseModel):
    id:int
    name:str
    department:str
    salary:Optional[float]=None
    is_active:Optional[bool]=True

employee=Employee(id='1',name="Soni",department="IT")
print(employee)
print(type(employee))
```

```python
employee=Employee(id='1',name="Soni",department="IT",salary=60000.123,is_active=False)
print(employee)
```

```python
from typing import List,Tuple,Dict
class Classroom(BaseModel):
    roomNumber:int
    students:List[str]
    capacity:int

classroom = Classroom(
    roomNumber=202,
    students=['soni','lakshmi','uday','nana'],
    capacity=4
)
print(classroom)
```

```python
classroom = Classroom(
    roomNumber=202,
    students={'soni':1,'lakshmi':2,'uday':3,'nana':4},
    capacity=4
)
print(classroom)
```

```python
from pydantic import ValidationError
try:
    classroom = Classroom(
    roomNumber=202,
    students={'soni':1,'lakshmi':2,'uday':3,'nana':4},
    capacity=4
    )
    print(classroom)
except ValueError as e:
    print("Validation Error")
    print(e)

    
```

```python
from pydantic import BaseModel

class Address(BaseModel):
    street:str
    city:str
    zip_code:int

class Customer(BaseModel):
    customer_id:int
    name:str
    address:Address

customer=Customer(
    customer_id=1,
    name="Soni",
    address={"street":"123 Main Street","city":"Nellore","zip_code":"12345"}
)

print(customer)
```

```python
from pydantic import BaseModel,Field
class Item(BaseModel):
    name:str=Field(min_length=4,max_length=15)
    price:float=Field(gt=0,le=1000)
    quantity:int=Field(ge=0)
item=Item(name="Soni",price=10,quantity=0)
print(item)
```

```python
item=Item(name="Soni",price=-1,quantity=0)
print(item)
```

```python
from pydantic import BaseModel,Field
class User(BaseModel):
    username:str=Field(description="Unique username for the user")
    age:int=Field(default=18,description="User age,defaults to 18")
    email:str=Field(default="user@example.com",description="Default email address")

user1=User(username="Soni")
print(user1)
user2=User(username="bob")
print(user2)

Field()
```
