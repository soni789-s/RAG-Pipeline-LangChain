```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
loader = TextLoader(r"C:\Users\udaya\OneDrive\Desktop\Langchain&Langgraph\Langchain_Hands_On\text.txt")
document = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000,chunk_overlap=30)
docs=text_splitter.split_documents(document)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model='nomic-embed-text')
db=FAISS.from_documents(docs,embeddings)
db
```

```python
query = "How does the speaker describe the desired outcome of war?"
docs=db.similarity_search(query)
docs[0].page_content
```

```python
retriever=db.as_retriever()
docs=retriever.invoke(query)
docs[0].page_content
```

```python
docs_and_score=db.similarity_search_with_score(query)
docs_and_score
```

```python
embedding_vector = embeddings.embed_query(query)
embedding_vector
```

```python
docs_score=db.similarity_search_by_vector(embedding_vector)
docs_score[0].page_content
```

```python
db.save_local("faiss_index")
```

```python
new_db=FAISS.load_local("faiss_index",embeddings,allow_dangerous_deserialization=True)
docs=new_db.similarity_search(query)
```

```python
docs
```

```python
from langchain_chroma import Chroma
vector_db = Chroma.from_documents(documents = docs,embedding =embeddings,persist_directory= "./chroma_db")
vector_db
```

```python
query_answer=vector_db.similarity_search(query)
query_answer[0].page_content
```

```python
db2 = Chroma(persist_directory="./chroma_db",embedding_function=embeddings)
docs=db2.similarity_search(query)
docs[0].page_content
```
